## Reading the Data Set

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv",encoding = "ISO-8859-1")
star_wars.head(10)
```

```python
star_wars.columns
```

## Cleaning Respondent ID column

```python
# removing any NaN row
star_wars = star_wars[star_wars["RespondentID"].notnull()]
star_wars.head(10)
```

## Cleaning Yes/No and NaN columns

```python
m = {
    "Yes":True,"No":False
}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(m)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(m)
star_wars.head(10)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
```

## Cleanining and Mapping Checkbox Columns

```python
movie_map = {
    'Star Wars: Episode I  The Phantom Menace': True,
    'Star Wars: Episode II  Attack of the Clones': True,
    'Star Wars: Episode III  Revenge of the Sith': True,
    'Star Wars: Episode IV  A New Hope': True,
    'Star Wars: Episode V The Empire Strikes Back': True,
    'Star Wars: Episode VI Return of the Jedi': True,
    np.NaN: False
}
```

```python
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_map)
star_wars.head(10)
```

## Renaming Columns

```python
col_name = {
    "Which of the following Star Wars films have you seen? Please select all that apply.":"seen_1",
    "Unnamed: 4":"seen_2",
    "Unnamed: 5":"seen_3",
    "Unnamed: 6":"seen_4",
    "Unnamed: 7":"seen_5",
    "Unnamed: 8":"seen_6"
}
```

```python
star_wars = star_wars.rename(col_name,axis=1)
```

```python
star_wars.head()
```

## Cleaning Rating Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
renaming_map = {
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
    "Unnamed: 10":"ranking_2",
    "Unnamed: 11":"ranking_3",
    "Unnamed: 12":"ranking_4",
    "Unnamed: 13":"ranking_5",
    "Unnamed: 14":"ranking_6"
}
```

```python
star_wars = star_wars.rename(renaming_map,axis=1)
star_wars.head()
```

## Calculating mean of rating columns

```python
mean_ratings = star_wars.iloc[:,9:15].mean()
mean_ratings
```

## Visualizing Ratings

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean_ratings.plot(kind="bar")
```

The ratings of 5th movie is lowest i.e for **Star Wars: Episode V The Empire Strikes Back**.

```python
mean_seen = star_wars.iloc[:,3:9].sum()
mean_seen
```

```python
mean_seen.plot(kind="bar")
```

5th movies i.e **Star Wars: Episode V The Empire Strikes Back** is the most seen movie.

## Exploring the Data by Binary Segementation

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

#### Response from Male Group

```python
mean_ratings = males.iloc[:,9:15].mean()
mean_ratings
```

```python
mean_ratings.plot(kind="bar")
```

Males liked the 5th Star Wars movie the most and hence it is ranked lowest.

```python
mean_seen = males.iloc[:,3:9].sum()
mean_seen
```

```python
mean_seen.plot(kind="bar")
```

Seen ratio of movie 5 is highest for males.

#### Response from Female Group

```python
mean_ratings = females.iloc[:,9:15].mean()
mean_ratings
```

```python
mean_ratings.plot(kind="bar")
```

Females also ranked 5th movie lowest.

```python
mean_seen = females.iloc[:,3:9].sum()
mean_seen
```

```python
mean_seen.plot(kind="bar")
```

Seen ratio for 5th movie is highest for females too.

## Conclusion :

So from above analysis we can conclude that the 5th Star Wars Movie i.e **Star Wars: Episode V The Empire Strikes Back** is Watched and Ranked highest by **both** males and females.
